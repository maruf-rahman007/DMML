```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dataset//iris.csv')
df.head()
df.describe()
```

```python
# Scatter Plot

x = df['sepal_length']
y = df['sepal_width']

x1 = df['petal_length']
y1 = df['petal_width']

plt.scatter(x, y, c ='r', label='Sepal Length an Width')
plt.scatter(x1, y1, c ='g', label='Petal Length an Width')

plt.title('Sepal vs Petal')
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.legend()
plt.show()
```
